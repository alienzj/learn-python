# Simple Line Plots

```python
%matplotlib inline 
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));
```

```python
plt.plot(x, np.sin(x));
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

```python
## Adjusting the Plot: Line Colors and Styles
```

```python
plt.plot(x, np.sin(x - 0), color='blue')                # specify color by name
plt.plot(x, np.sin(x - 1), color='g')                   # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')                # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')             # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3))       # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse');  
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the follow ing codes:
plt.plot(x, x + 6, linestyle='-')
plt.plot(x, x + 7, linestyle='--')
plt.plot(x, x + 8, linestyle='-.')
plt.plot(x, x + 9, linestyle=':');
```

```python
# These linstyle and color codes can be combined into a single non-keyword argument

plt.plot(x, x + 0, '-g')
plt.plot(x, x + 2, '--c')
plt.plot(x, x + 4, '-.k')
plt.plot(x, x + 6, ':r');
```

```python
## Adjusting the Plot: Axes Limits
```

```python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

```python
plt.plot(x, np.sin(x))

plt.xlim(10, 0)
plt.ylim(1.2, -1.2);
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]);  # [xmin, xmax, ymin, ymax]
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight');
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal')
```

## Labeling Plots
- titles
- axis labels
- simple legends

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend();
```

## Aside: Matplotlib Gotchas
- plt.plot() -> ax.plot()
- plt.legend() -> ax.legend()
- plt.xlabel() -> ax.set_xlabel()
- plt.ylabel() -> ax.set_ylabel()
- plt.xlim() -> ax.set_xlim()
- plt.ylim() -> ax.set_ylim()
- plt.title() -> ax.set_title()

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
       xlabel='x', ylabel='sin(x)',
       title='A Simple Plot')
```
