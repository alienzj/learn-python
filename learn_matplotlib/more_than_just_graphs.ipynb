```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1, aspect=1)
plt.grid()
plt.show()
```

```python
ax.scatter([.5], [.5], c='yellow', edgecolor='gold',
           linewidth=15, s=130000)
plt.grid()
plt.show()
```

```python
ax.scatter([.5], [.5], c='yellow', edgecolor='gold',
           linewidth=15, s=130000)
ax.scatter([.37], [.60], c='darkgoldenrod', s=1700)
plt.grid()
plt.show()
```

```python
X = np.linspace(.37, .63, 100)
Y = 4 * (X - .5) ** 2 + 0.3
ax.plot(X, Y, c='darkgoldenrod', linewidth=20)

plt.grid()
plt.show()
```

```python
a = np.linspace(.6, .7, 100)
b = -1.75 * (a - .7) ** 2 + 0.6

c = np.linspace(.63, .75, 100)
d = -1.25 * ( a - .6) ** 1.5 + 0.7

e = np.linspace(.25, .35, 100)
f = -2 * (a - .8) ** 2 + 0.77

ax.plot(a, b, c='darkgoldenrod', linewidth=11)
ax.plot(c, d, c='darkgoldenrod', linewidth=9)
ax.plot(e, f, c='darkgoldenrod', linewidth=9)

plt.grid()
plt.show()
```

```python
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.set_xticks([])
ax.set_yticks([])

plt.grid()
plt.show()
```
